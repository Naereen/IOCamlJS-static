# H.261 Video Decoding in OCaml.

This notebook will demonstrate a simple H.261 Video Decoder written in OCaml and compiled to javascript with js_of_ocaml.  Along with it a simple support library for video coding will also be described.  As yet these projects are at an early stage, and are mainly done as a bit of fun.

* [Ovideo](https://github.com/ujamjar/ovideo) - video compression utilities (dct, bitstream, yuvframe etc)
* [Oh261](https://github.com/ujamjar/oh261) - h.261 decoder (and an incomplete encoder)

```ocaml
open Ovideo
open Oh261
```

### Setting up the bitstream

This decoder needs a module of type Ovideo.Bits.Source to supply the bitstream.  We will use an (extended) XMLHttpRequest to return the bitstream as a uint8 typed array.

```ocaml
(* the bitstream sourced from a typed array *)
module Typed_array_source = struct
  type t = int ref * int * Typed_array.uint8Array Js.t
  let init b = ref 0, b##length, b
  let next_byte (pos,len,bytes) = 
    if !pos < len then
      let b = Typed_array.unsafe_get bytes !pos in
      pos := !pos + 1;
      Some b
    else
      None
end
```

This is a small hack to do a XMLHttpRequest and return a typed array.  Actually, return isn't quite right - the call is asynchronous and thus the data is only available in the callback passed to `load_url`.

```ocaml
(* extend the standard js_of_ocaml xmlHttpRequest object to 
   support returning TypedArray's *)
class type xmlHttpRequest2 = object
    inherit XmlHttpRequest.xmlHttpRequest
    method response : Typed_array.arrayBuffer Js.t Js.prop
    method responseType : Js.js_string Js.t Js.prop
end

(* load data from the url asyncronously.  data processing must
   therefore happen within the callback function *)
let load_url url f = 
    let xml : xmlHttpRequest2 Js.t = jsnew (Js.Unsafe.global##_XMLHttpRequest)() in
    let () = xml##_open(Js.string "GET", Js.string url, Js._true) in
    let () = xml##responseType <- Js.string "arraybuffer" in
    let () = xml##send(Js.null) in
    xml##onreadystatechange <- Js.wrap_callback (
        fun _ ->
            match xml##readyState with
            | XmlHttpRequest.DONE -> 
                f (jsnew (Typed_array.uint8Array_fromBuffer)(xml##response))
            | _ -> ())
```

Now we have defined the bits source we can construct the bitstream reader module.

```ocaml
module R = Ovideo.Bits.Reader(Typed_array_source)
```

### YUV video frames

Video frames are represented as 3 planes of unsigned 8 bit ints.  The luma plane is the full size of the video frame.  The 2 chroma planes are subsampled by a factor of 2 both horizontally and vertically.

```ocaml
module F = Ovideo.Frame.U8
module P = F.Plane
```

### Setting up the decoder module

This builds the decoder working over our bitstream source.

```ocaml
module H261 = Oh261.Decode.Make(R)
```

### Displaying the output frame.

To display the video we will use a canvas element.  A canvas requires the YUV planes to be converted to RGB.

Note that 4 pixels are processed per iteration as they all share the same chroma values (due to the subsampling).  Also, for performance reasons, bigarray accesses are converted from 2d to 1d which is a lot faster.

```ocaml
let convert_to_rgb width height f img = 
  (* 1d array accesses are much faster than 2d *)
  let y' = P.to_1d f.F.y in
  let u' = P.to_1d f.F.u in
  let v' = P.to_1d f.F.v in

  let oy' = ref 0 in
  let ouv' = ref 0 in
  let rgb' = ref 0 in

  for y=0 to height/2-1 do
    for x=0 to width/2-1 do

      let oy, ouv = !oy', !ouv' in

      let y0' = y'.{oy} in
      let y1' = y'.{oy+1} in
      let y2' = y'.{oy+width} in
      let y3' = y'.{oy+width+1} in

      let u' = u'.{ouv} in
      let v' = v'.{ouv} in

      oy' := !oy' + 2;
      ouv' := !ouv' + 1;

      (* approx conversion in integer arithmetic *)
      let r' = (v' + ((v' * 103) lsr 8)) - 179 in
      let g' = ((u' * 88) lsr 8) - 44 + ((v' * 183) lsr 8) - 91 in
      let b' = (u' + ((u' * 198) lsr 8)) - 227 in
     
      let rgb = !rgb' in
      Dom_html.pixel_set img (rgb+0) (y0' + r');
      Dom_html.pixel_set img (rgb+1) (y0' - g');
      Dom_html.pixel_set img (rgb+2) (y0' + b');
      Dom_html.pixel_set img (rgb+3) 255;                   
      let rgb = rgb+4 in                                    
      Dom_html.pixel_set img (rgb+0) (y1' + r');
      Dom_html.pixel_set img (rgb+1) (y1' - g');
      Dom_html.pixel_set img (rgb+2) (y1' + b');
      Dom_html.pixel_set img (rgb+3) 255;                   
      let rgb = rgb+(width*4) in                            
      Dom_html.pixel_set img (rgb+0) (y3' + r');
      Dom_html.pixel_set img (rgb+1) (y3' - g');
      Dom_html.pixel_set img (rgb+2) (y3' + b');
      Dom_html.pixel_set img (rgb+3) 255;                   
      let rgb = rgb-4 in                                    
      Dom_html.pixel_set img (rgb+0) (y2' + r');
      Dom_html.pixel_set img (rgb+1) (y2' - g');
      Dom_html.pixel_set img (rgb+2) (y2' + b');
      Dom_html.pixel_set img (rgb+3) 255;

      rgb' := !rgb' + 8
    done;

    oy' := !oy' + width;
    rgb' := !rgb' + (width*4);

  done
```

This initializes the canvas element.

```ocaml
(* get an element from the page with the given id *)
let get_element id = 
    Js.Opt.get (Dom_html.document##getElementById (Js.string id)) (fun () -> assert false)

let jstri i = Js.string (string_of_int i)

(* create the canvas *)
let create_canvas width height = 
    let d = Dom_html.document in
    let canvas = Dom_html.createCanvas d in
    let style = canvas##style in
    canvas##width <- width;
    canvas##height <- height;
    style##width <- jstri width;
    style##height <- jstri height;
    canvas

(* attach the canvas to the given dom element *)
let attach_canvas width height id = 
  let div = get_element id in
  let canvas = create_canvas width height in
  Dom.appendChild div canvas;
  canvas
```

### Decoding

```ocaml
let display width height cur ctx img = 
    convert_to_rgb width height cur img##data;
    ctx##putImageData(img, 0., 0.)

let rec decode width height h261 ctx img =
    try
        (* run decoder *)
        H261.decode_picture h261;
        (* convert and display *)
        display width height h261.H261.State.cur ctx img;
        (* loop *)
        ignore (Dom_html.window##setTimeout(
                    Js.wrap_callback (fun () -> decode width height h261 ctx img), 
                    0.0))
    with exc -> 
        Firebug.console##log(Js.string (Printexc.to_string exc))
```

### Running the decoder

Finally, we have our decoder.  It loads the bitstream from the given url, creates the rendering canvas and does the decoding.

```ocaml
let run_decoder width height id url = 
    load_url url (fun bitstream ->
        (* setup the bitstream *)
        let bits = R.init (Typed_array_source.init bitstream) in
        (* setup the h.261 decoder *)
        let h261 = H261.State.init bits in
        (* create the canvas, get the context and image data*)
        let canvas = attach_canvas width height id in
        let ctx = canvas##getContext(Dom_html._2d_) in
        let img = ctx##createImageData(width, height) in
        decode width height h261 ctx img
    )
```

We now play a little trick so we can embed the decoder in the output area of the cell when its run.  The idea is we attach the decoder function to the `Dom` window object, then call it from a script which is embedded in the output cell.

```ocaml
class type global = object
    method runDecoder : (unit -> unit) Js.prop
end
let run =
    let id = ref 0 in
    (fun width height url ->
        (* create a unique id for the div *)
        incr id;
        let id = "id" ^ string_of_int !id in
        (* install our decode function *)
        (Js.Unsafe.global : global Js.t)##runDecoder <- 
            (fun () -> run_decoder width height id url);
        (* embed the player in the output cell *)
        Iocaml.display "text/html" ("
            <div id=\""^id^"\"></div>
            <script>window.runDecoder()</script>
        ")
    )
```
