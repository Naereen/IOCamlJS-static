```ocaml
module type Show = sig
    type t
    val show : t -> unit
end
```

```ocaml
let print (implicit S : Show) x = S.show x
```

```ocaml
implicit module Show_int = struct
    type t = int
    let show x = Printf.printf "Show_int: %i\n" x
end
```

```ocaml
implicit module Show_float = struct
    type t = float
    let show x = Printf.printf "Show_float: %f\n" x
end
```

```ocaml
implicit functor Show_list (X : Show) = struct
        type t = X.t list
        let show x =
            Printf.printf "Show_list: [\n";
            List.iter X.show x;
            Printf.printf "]\n"
    end
```
